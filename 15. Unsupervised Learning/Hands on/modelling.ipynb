```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_df = scaler.fit_transform(df)
```

```python
scaled_df = pd.DataFrame(scaled_df, columns= df.columns)
scaled_df
```

```python
inertia = []

range_val = range(2,15)
for i in range_val:
    kmean = KMeans(n_clusters=i, random_state=10)
    kmean.fit(scaled_df)
    pred = kmean.predict(scaled_df)
    inertia.append(kmean.inertia_)
    
plt.plot(range_val, inertia, 'bx-') 
plt.xlabel('k-values')
plt.title("Elbow method to find the optimal k value")
plt.show()
    
```

```python
model = KMeans(4, random_state=10)
model.fit(scaled_df)
```

```python
model.predict(scaled_df)
```
