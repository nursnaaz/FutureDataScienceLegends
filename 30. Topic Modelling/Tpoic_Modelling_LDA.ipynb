```python
import nltk
import gensim
import pyLDAvis
import pyLDAvis.gensim_models
from nltk.tokenize import word_tokenize
import gensim.corpora as corpora
from gensim.models.ldamodel import LdaModel

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
from nltk.corpus import stopwords
```

```python
document = [
    'The bank called for money',
    'The bank said the money was approved',
    'The boat was on the river',
    'She deposited money in the bank',
    'The approved loan was transfered',
    'The river is following very aggrasively',
    'The river water is very clean',
    'The river bank is filled with water',
    'The river bank is full of water',
    'The people from river bank asked to vacate'
]

```

```python
stop_word = set(stopwords.words('english'))
```

```python
preprocessed_token = []
clean_document = []
for doc in document:
  doc = doc.lower()
  print(word_tokenize(doc))
  tokens = word_tokenize(doc)
  for word in tokens:
    if word not in stop_word:
      if word.isalpha():
        preprocessed_token.append(word)
  clean_document.append(preprocessed_token)
  preprocessed_token = []




```

```python
# Build LDA model
lda_model = LdaModel(corpus=corpus,
                     id2word=dictionary,
                     num_topics=20, # You can adjust the number of topics
                     update_every=1,
                     passes=100,
                     alpha='auto')

# Print the topics
topics = lda_model.print_topics(num_words=10) # You can adjust the number of words per topic
for topic in topics:
    print(topic)

# Visualize the topics
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim_models.prepare(lda_model, corpus, dictionary)
pyLDAvis.display(vis)
```

```python
# prompt: with this lda_model object can you write a script to predict a following news which topic it falls under = 'Russian air defences down dozens of Ukrainian drones, including two near St. Petersburg
# Ukraine Governors report downing drones in Russia, Ukraine conflict; Russian forces target Ukrainian cities with drones and missiles
# Published - July 05, 2025 07:12 am IST
# AP, Reuters
# Birds fly over a residential building as smoke rises from the city after a Russian drone and missile strike, amid Russia‚Äôs attack on Ukraine, in Kyiv, Ukraine, on July 4, 2025'

def predict_news_topic(lda_model, dictionary, news_text):
  """
  Predicts the topic of a given news article using a trained LDA model.

  Args:
    lda_model: Trained Gensim LdaModel object.
    dictionary: Gensim Dictionary used to create the corpus for training.
    news_text: The text of the news article to predict the topic for.

  Returns:
    A tuple containing the most probable topic index and its probability,
    or None if no words from the news text are in the dictionary.
  """
  dictionary.filter_extremes(no_below=5, no_above=0.5)
  stop_word = set(stopwords.words('english'))
  processed_text = []

  # Preprocess the news text
  news_text = news_text.lower()
  tokens = word_tokenize(news_text)
  for word in tokens:
    if word not in stop_word and word.isalpha():
      processed_text.append(word)

  if not processed_text:
    return None # Return None if no words remain after preprocessing

  # Create a bag-of-words representation for the news text
  bow_news = dictionary.doc2bow(processed_text)

  if not bow_news:
      return None # Return None if no words from the text are in the dictionary

  # Get the topic distribution for the news text
  topic_distribution = lda_model.get_document_topics(bow_news)

  # Find the topic with the highest probability
  if topic_distribution:
      highest_topic = max(topic_distribution, key=lambda item: item[1])
      return highest_topic
  else:
      return None


news_article = 'Russian air defences down dozens of Ukrainian drones, including two near St. Petersburg\nUkraine Governors report downing drones in Russia, Ukraine conflict; Russian forces target Ukrainian cities with drones and missiles\nPublished - July 05, 2025 07:12 am IST\nAP, Reuters\nBirds fly over a residential building as smoke rises from the city after a Russian drone and missile strike, amid Russia‚Äôs attack on Ukraine, in Kyiv, Ukraine, on July 4, 2025'

predicted_topic = predict_news_topic(lda_model, dictionary, news_article)

if predicted_topic:
  print(f"The news article most likely belongs to Topic {predicted_topic[0]} with probability {predicted_topic[1]:.4f}")
  # You can also print the top words for that topic to understand its theme
  print("Top words for this topic:")
  print(lda_model.print_topic(predicted_topic[0], topn=10))
else:
  print("Could not determine the topic for the news article (possibly no relevant words found).")
```
