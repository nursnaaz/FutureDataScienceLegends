# Plots

```python
# scatter plot - relationship between variables
```

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [10, 15, 13, 18, 20]

plt.scatter(x,y)

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('scatter plot sample')
plt.show()
```

```python
# bar chart - comparison of a variable withitself
```

```python
categories = ['cat-A', 'cat-b', 'cat-c']
values = [15,30,25]

plt.bar(categories, values)
plt.show()
```

```python
# histogram - to see distribution of values
```

```python
data = [1,2,2,3,3,3,4,4,5,5,5,5]

"""
b1 = 3
1 = 1 times occurred
2 = 2 times occurred

b2 = 9
3 = 3 times occurred
4 = 2 times occurred
5 = 4 times occurred
"""

plt.hist(data)
plt.show()
```

```python
plt.hist(data, bins=2)
plt.show()

# bins = 2
# b1 = 1-3
# b2 = 3-5
```

```python
b1 = 3
b2 = 9
```

```python
plt.hist(data, bins=3, edgecolor='red')
plt.show()
```

```python
"""
b1 = 3
1 = 1 times occurred
2 = 2 times occurred

b2 = 3
3 = 3 times occurred

b3 = 6
4 = 2 times occurred
5 = 4 times occurred
```

```python
# pie chart - to see composition of a group on the whole data
```

```python
catg = ['catA', 'catB', 'catC']
sizes = [15,10,25]

# catA = 15 --> 15/50= 0.3 = 30%
# catB = 10 --> 10/50= 0.2 = 20%
# catC = 25 --> 25/50= 0.5 = 50%
# total = 50  -> 100%
plt.pie(sizes, labels=catg, autopct="%.1f%%")
plt.show()
```

```python
#boxplot = to find the outlier

data = [85, 90, 70, 100, 110, 95, 75, 120, 300, 100, 200]

plt.boxplot(data)
plt.show()
```
