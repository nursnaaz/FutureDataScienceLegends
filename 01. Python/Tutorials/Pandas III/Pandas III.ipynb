#### Requirement: Show the top 10 locations list having the fast coasters(take avg of multiple coasters)

```python
df.Location.value_counts()
```

```python
df.query('Location != "Other"')['Location'].value_counts()
```

```python
df.query('Location != "Other"')[['Location','Speed_mph']].value_counts()
```

```python
df.query('Location != "Other"').dropna(subset=['Speed_mph']).groupby('Location')['Speed_mph'].agg(['mean']).sort_values(by='mean', ascending=False).head(10)
```

```python
# Requirement: Show the top 10 locations list having the fast coasters(take avg of multiple coasters)

# the same muliple methods but in multiple line
df.query('Location != "Other"') \
    .dropna(subset=['Speed_mph']) \
    .groupby('Location')['Speed_mph'] \
    .agg(['mean']) \
    .sort_values(by='mean', ascending=False)\
    .head(10)\
    .plot(kind='barh')
```

```python
# muliple methods in single line
df.query('Location != "Other"').dropna(subset=['Speed_mph']).groupby('Location')['Speed_mph'].agg(['mean']).sort_values(by='mean', ascending=False).head(10).plot(kind='barh')
```
