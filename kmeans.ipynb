```python
import numpy as np
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
# create data points
X= np.array([[1,2],[1,4],[1,0],
             [10,2],[10,4],[10,0]])
X
```

```python
plt.figure(figsize=(3,3))
plt.scatter(X[:,0],X[:,1])
```

```python
from sklearn.cluster import KMeans
```

```python
Kmeans=KMeans(n_clusters=3,random_state=42)
Kmeans.fit(X)
```

```python
pred=Kmeans.predict(X)
pred
```

```python
import pandas as pd
```

```python
df_X=pd.DataFrame(X)
df_X.columns=['X1','X2']
df_X
```

```python
df_y = pd.DataFrame(pred)
df_y.columns=['cluster label']
df_y
```

```python
for i in range (1,6):
  kmeans=KMeans(n_clusters = i, random_state=42) # Call KMeans (the class) not kmeans (the variable)
  kmeans.fit(X)
  print(f'k={i}')
  print(kmeans.cluster_centers_)
  print('*'*15)
  # Access cluster centers using cluster_centers_ only
  # print(kmeans.cluster_centres_)  # This attribute is likely deprecated/incorrect
  print('*'*15)
```

```python
fig=plt.figure(figsize=(10,3))

ax1=fig.add_subplot(131)
ax1.scatter(X[:,0],X[:,1])
ax1.scatter(5.5,2)
ax1.set_title('k=1')

ax1 =fig.add_subplot(132)
ax1.scatter(X[:,0],X[:,1],marker='+',c='black' , s=100)
ax1.scatter(1,2, marker='o',c='cyan')
ax1.scatter(10,2, marker='o',c='cyan')
ax1.set_title('k=2')

ax1 =fig.add_subplot(133)
ax1.scatter(X[:,0],X[:,1],marker='+',c='black' , s=100)
ax1.scatter(1,2, marker='o',c='cyan')
ax1.scatter(10,2, marker='o',c='cyan')
ax1.scatter(1,4, marker='o',c='cyan')
ax1.set_title('k=3')


plt.show()

```

```python
res=[]
for i in range (1,6):
  kmeans=KMeans(n_clusters = i, random_state=42)
  kmeans.fit(X)
  res.append(kmeans.inertia_)
```

```python
plt.plot(range(1,6),res )
plt.show()
```

```python
# optimal k=2
kmeans=KMeans(n_clusters = 2, random_state=42)
kmeans.fit(X)
```
