```python
from keras.datasets import mnist
import matplotlib.pyplot as plt
from keras.utils import to_categorical
from keras_visualizer import visualizer
from keras.initializers import HeNormal


from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, BatchNormalization
from keras.optimizers import Adam
from keras.losses import CategoricalCrossentropy
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
for i in range(1, 5):
    # Get the current example
    example = x_train[i]

    # Print each row of the example on its own line
    for row in example:
        print('[' + ' '.join(f"{val:3d}" for val in row) + ']')

    # Add separator between examples
    print('-' * 80)

    plt.imshow(x_train[i],cmap='gray')
    print("Y - data : ",y_train[i])
    plt.show()
```

```python
Y_train = to_categorical(y_train)
Y_test = to_categorical(y_test)
```

```python
x_train.shape[1], x_train.shape[2], 1
```

```python
x_train.shape
```

```python
# CNN Architecture
initializer = HeNormal()
```

```python
model = Sequential()

#2^n
```

```python
model.add(Conv2D(input_shape = (x_train.shape[1], x_train.shape[2], 1), filters = 64, kernel_size= (3,3), strides = (2,2),
                 padding = 'valid', activation = 'relu', kernel_initializer = initializer))
model.add(MaxPooling2D(pool_size = (2,2)))
model.add(Flatten())
model.add(Dense(units = 100, activation='relu', kernel_initializer = initializer))
model.add(BatchNormalization())
model.add(Dense(units = 10, activation='softmax', kernel_initializer = initializer))
```

```python
adam_optimizer = Adam()
category_loss = CategoricalCrossentropy()
model.compile(optimizer = adam_optimizer, loss = category_loss, metrics = ['accuracy'])
```

```python
history = model.fit(x_train, Y_train, epochs=19, batch_size=10, validation_data=(x_test, Y_test) ,verbose=1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

# Load the image
img_path = 'six.png'  # Replace with actual path
img = cv2.imread(img_path, cv2.IMREAD_UNCHANGED)  # Read with transparency

# Handle transparency properly
if img.shape[-1] == 4:  # If has alpha channel
    # Create a white background
    background = np.ones((img.shape[0], img.shape[1], 3), dtype=np.uint8) * 255

    # Normalize alpha channel to range 0-1
    alpha = img[:,:,3] / 255.0

    # Get foreground
    for c in range(3):
        background[:,:,c] = (1-alpha) * background[:,:,c] + alpha * img[:,:,c]

    img = background.astype(np.uint8)

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Since this is a digit on transparent background made black on white,
# we need to invert to match MNIST style (white digit on black background)
gray = cv2.bitwise_not(gray)

# Resize to match expected input dimensions
resized = cv2.resize(gray, (28, 28))

# Convert to float32 to match training data type
resized = resized.astype('float32')

# Debug: Show the processed image
plt.figure(figsize=(4, 4))
plt.imshow(resized, cmap='gray')
plt.title('Processed Image')
plt.axis('off')
plt.show()

# Reshape according to model's expected input shape
input_img = np.expand_dims(resized, axis=0)
input_img = np.expand_dims(input_img, axis=-1)

# Make prediction
prediction = model.predict(input_img)
predicted_class = np.argmax(prediction, axis=1)[0]
confidence = np.max(prediction) * 100

# Display results
print(f"Predicted digit: {predicted_class}")
print(f"Confidence: {confidence:.2f}%")

# Display the probabilities for all classes
plt.figure(figsize=(10, 4))
plt.bar(range(10), prediction[0])
plt.xticks(range(10))
plt.xlabel('Digit')
plt.ylabel('Probability')
plt.title('Prediction Probabilities')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from PIL import Image
from io import BytesIO
import requests

# Load the image
img_path = 'five.jpeg'  # Replace with actual path
img = cv2.imread(img_path, cv2.IMREAD_UNCHANGED)

# Display original image
plt.figure(figsize=(4, 4))
if img.shape[-1] == 4:  # If has alpha channel
    # Convert RGBA to RGB for display
    plt.imshow(cv2.cvtColor(img[:, :, :3], cv2.COLOR_BGR2RGB))
else:
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')
plt.show()

# Convert to grayscale, handling transparency if present
if img.shape[-1] == 4:  # If has alpha channel
    # Extract the alpha channel
    alpha = img[:,:,3]
    # Create a binary mask (0 = transparent, 255 = opaque)
    mask = alpha.copy()
    mask[mask > 0] = 255
    # Use the mask to get just the digit (black on white background)
    gray = np.ones_like(mask) * 255
    gray[mask > 0] = 0
else:
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    # Ensure black digit on white background
    if np.mean(gray) < 128:
        gray = cv2.bitwise_not(gray)

# For debugging, display the grayscale version
plt.figure(figsize=(4, 4))
plt.imshow(gray, cmap='gray')
plt.title('Grayscale Image (Pre-processing)')
plt.axis('off')
plt.show()

# Now the crucial part - convert to MNIST format
# MNIST has white digit on black background
gray = cv2.bitwise_not(gray)

# Crop to content (remove unnecessary whitespace)
# Find all non-zero points (the digit)
coords = cv2.findNonZero(gray)
x, y, w, h = cv2.boundingRect(coords)
# Add small padding
padding = 2
x = max(0, x - padding)
y = max(0, y - padding)
w = min(gray.shape[1] - x, w + 2*padding)
h = min(gray.shape[0] - y, h + 2*padding)
# Crop to the content
digit = gray[y:y+h, x:x+w]

# Resize to 20x20 (standard for MNIST)
digit_resized = cv2.resize(digit, (20, 20), interpolation=cv2.INTER_AREA)

# Create 28x28 canvas (standard for MNIST)
canvas = np.zeros((28, 28), dtype=np.float32)
# Place the digit on the canvas with some padding (as in MNIST)
offset_x = (28 - 20) // 2
offset_y = (28 - 20) // 2
canvas[offset_y:offset_y+20, offset_x:offset_x+20] = digit_resized

# Convert to float32 (same as your model input)
canvas = canvas.astype('float32')

# Display the final processed image
plt.figure(figsize=(4, 4))
plt.imshow(canvas, cmap='gray')
plt.title('Processed Image (MNIST Format)')
plt.axis('off')
plt.show()

# Reshape for model input
input_img = np.expand_dims(canvas, axis=0)
input_img = np.expand_dims(input_img, axis=-1)

# Print shape information
print("Input shape:", input_img.shape)
print("Min pixel value:", np.min(input_img))
print("Max pixel value:", np.max(input_img))

# Make prediction
prediction = model.predict(input_img)
predicted_class = np.argmax(prediction, axis=1)[0]
confidence = np.max(prediction) * 100

# Display results
print(f"Predicted digit: {predicted_class}")
print(f"Confidence: {confidence:.2f}%")

# Display the probabilities for all classes
plt.figure(figsize=(10, 4))
plt.bar(range(10), prediction[0])
plt.xticks(range(10))
plt.xlabel('Digit')
plt.ylabel('Probability')
plt.title('Prediction Probabilities')
plt.show()
```
