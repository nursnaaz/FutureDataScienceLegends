```python
iteration = 100000
learning_rate = 0.001
m = 0
b = 0
error_list = []
error_sum = 0
for i in range(iteration):
    print("input : ",X_)
    y_pred = m * X_ + b
    print("m = ",m)
    print("b = ",b)
    print("output prediction :",y_pred)
    print("Actual output : ",y)
    old_error = error_sum
    error = y_pred - y
    print("Error : ",error)
    print("Sum of Error : ",abs(np.sum(error)))
    error_sum = abs(np.sum(error))
    error_list.append(abs(np.sum(error)))
    m_gradient = (1/n) * np.dot((error), X_)
    b_gradient = (1/n) * np.sum((error))
    m = m - (m_gradient * learning_rate)
    b = b - (b_gradient * learning_rate)
    print(m)
    print(b)
    print(old_error)
    change = abs(error_sum  - old_error)
    if change < 0.001:
        break
```
