```python
def equ(x):
    return 6*x*x + 2*x + 2
```

```python
equ(0)
```

```python
y = lambda x: 6*x*x + 2*x + 2
```

```python
y(0)
```

```python
der = lambda x : 12*x + 2
```

```python
der(-1)
```

```python
learning_rate = 0.01
```

```python
cur_x = -1
```

```python
num_of_iteration = 100
```

```python
change = 0
```

```python
while num_of_iteration>=0:
    prev_x = cur_x
    cur_x = cur_x - der(cur_x) * 0.01
    change = prev_x - cur_x
    print("Value of y equation : ",y(cur_x))
    print("Value of x in y :",cur_x)
    print("change of x in y :",change)
    num_of_iteration-=1
```

```python
while num_of_iteration>=0:
    prev_x = cur_x
    cur_x = cur_x - der(cur_x) * learning_rate
    change = abs(prev_x - cur_x)
    print("Value of y equation : ",y(cur_x))
    print("Value of x in y :",cur_x)
    print("change of x in y :",change)
    num_of_iteration-=1
    if change<0.001:
        break
```
