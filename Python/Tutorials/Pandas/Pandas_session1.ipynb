```python
import pandas as pd
```

##### Load data into series from lists

```python
data1 = [101,102,103]
print(data1, type(data1))
```

```python
field1 = pd.Series(data1)
print(field1, type(field1))
```

```python
data2 = ['Sree', 'Sai', 'Kishore']
print(data2, type(data2))
```

```python
field2 = pd.Series(data2)
print(field2, type(field2))
```

##### Load data into dataframe from lists, series, dictionaries

```python
df1 = pd.DataFrame({'eid':field1, 'ename': field2})
df1
```

```python
df1 = pd.DataFrame({'eid':data1, 'ename': data2})
df1
```

```python
data1 = {'eid': [101,102,103], 'ename': ['Sree', 'Sai', 'Kishore']}
print(data1, type(data1))
```

```python
df1 = pd.DataFrame(data1)
df1
```

##### assign indices to the rows

```python
df1 = pd.DataFrame(data1, index=['r1','r2','r3'])
df1
```
